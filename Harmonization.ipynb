```python
# This one return dictionary

def read_sensor_data(csv_list):
    # csv_list = glob.glob('Lauren purpleair data\\*\\*.csv')
    numeric_columns = ['time_stamp', 'temperature', 'humidity', 'pm2.5_alt', 'pressure']
    sensor_dataframes = {}

    for idx, csv_path in enumerate(csv_list):
        sensor_df = pd.read_csv(csv_path)

        # If the file has fewer columns, tryaa different delimiter
        if len(sensor_df.columns) < 4:
            sensor_df = pd.read_csv(csv_path, delimiter=';')

        # Dynamically check for 'Sensor Name' and 'sensor_index' columns
        extended_columns = numeric_columns.copy()
        if 'Sensor Name' in sensor_df.columns:
            extended_columns.append('Sensor Name')
        if 'sensor_index' in sensor_df.columns:
            extended_columns.append('sensor_index')

        # Append the required columns to the list of dataframes
        # sensor_dataframes.append(sensor_df[extended_columns])


        sensor_dataframes[csv_path.split('/')[-2]] = sensor_df[extended_columns]

    return sensor_dataframes


def get_sensor_dataframes(path_1_120, path_121_206):

    sensor_names_1_120, sensor_names_121_206 = [], []
    # for 1 - 120
    csv_list = glob.glob(path_1_120)
    sensor_dataframes_1_120 = read_sensor_data(csv_list) #[:10]
    # sensor_names_1_120 = [csv_path.split('/')[-2] for csv_path in csv_list] #[:10]



    # # # for 121 - 206
    sensor_dataframes_121_206 = read_sensor_data([path_121_206])['Data'] #[:10]
    sensor_names_121_206 = list(set(sensor_dataframes_121_206['Sensor Name'].to_list()))

    # Convert the "time_stamp" column to datetime objects
    sensor_dataframes_121_206["time_stamp"] = pd.to_datetime(sensor_dataframes_121_206["time_stamp"], unit="s")
    # Format the datetime objects as ISO 8601 strings with "Z" for UTC
    sensor_dataframes_121_206["time_stamp"] = sensor_dataframes_121_206["time_stamp"].dt.strftime("%Y-%m-%dT%H:%M:%SZ")
    sensor_dataframes_121_206 = sensor_dataframes_121_206.rename(columns={"timestamp": "time_stamp"})
    # Group the DataFrame by "sensor_index"
    grouped_df = sensor_dataframes_121_206.groupby("sensor_index")

    # Create separate DataFrames for each sensor_index
    sensor_dfs = {}
    for sensor_index, group in grouped_df:
        # sensor_dfs.append(group.reset_index(drop=True))
        df = group.reset_index(drop=True)
        sensor_dfs[df['Sensor Name'][0]] = df
    sensor_dataframes_121_206 = sensor_dfs.copy()

    return sensor_dataframes_1_120, sensor_dataframes_121_206, sensor_names_1_120, sensor_names_121_206


path_1_120 = 'Data/purpleair data//*//*.csv'
path_121_206 = 'Data/sensor_history_121_206_10_mins.csv'

sensor_dataframes_1_120, sensor_dataframes_121_206, sensor_names_1_120, sensor_names_121_206 = get_sensor_dataframes(path_1_120, path_121_206)
```

```python
import pandas as pd
from datetime import datetime

def clean_sensor_dataframes(sensor_dataframes,flag_1_120=True):

    numeric_columns = ['temperature','humidity','pm2.5_alt','pressure']

    cleaned_dataframes = {}

    # for idx, sensor_df in enumerate(sensor_dataframes):
    for key, sensor_df in sensor_dataframes.items():
        # Remove the header row and create a copy of the DataFrame
        sensor_df = sensor_df.iloc[1:].copy()

        # Convert 'time_stamp' to datetime (and ensure timezone-naive)
        sensor_df['time_stamp'] = pd.to_datetime(sensor_df['time_stamp']).dt.tz_localize(None)

        if flag_1_120:
            start_date = datetime(2023, 6, 14)
            end_date = datetime(2023, 6, 27)
            sensor_df = sensor_df[(sensor_df['time_stamp'] >= start_date) & (sensor_df['time_stamp'] <= end_date)].copy()
      
        # Convert specified columns to numeric, handling invalid parsing
        sensor_df[numeric_columns] = sensor_df[numeric_columns].apply(pd.to_numeric, errors='coerce')

        sensor_df = sensor_df.dropna()
        # Append the cleaned DataFrame to the list

        # sort by time_stamp
        sensor_df = sensor_df.sort_values(by='time_stamp').reset_index(drop=True)
        
        # cleaned_dataframes.append(sensor_df)
        cleaned_dataframes[key] = sensor_df

    return cleaned_dataframes


cleaned_sensor_dataframes_1_120 = clean_sensor_dataframes(sensor_dataframes_1_120,flag_1_120=True)
cleaned_sensor_dataframes_121_206 = clean_sensor_dataframes(sensor_dataframes_121_206,flag_1_120=False)
```

```python
def aggregate_sensor_data(sensor_dataframes, column_name):
    """
    Aggregates the specified column from all sensor DataFrames, aligning rows based on `time_stamp`.
    """
    # Initialize the combined DataFrame with the `time_stamp` column from the first sensor
    combined_df = pd.DataFrame(sensor_dataframes[next(iter(sensor_dataframes))]['time_stamp'])
    combined_df = combined_df.rename(columns={'time_stamp': 'time_stamp'})

    for key, df in sensor_dataframes.items():
        # Merge each sensor's data on `time_stamp`
        combined_df = pd.merge(
            combined_df,
            df[['time_stamp', column_name]].rename(columns={column_name: key}),
            on='time_stamp',
            how='inner'  # Use 'inner' join to keep only overlapping timestamps
        )
        # print(len(combined_df),key)
    # del combined_df['time_stamp']
    return combined_df
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_rmse(sensor_data, rolling_median):
    """
    Calculates the Root Mean Squared Error (RMSE) between sensor data and rolling median.
    """
    return np.sqrt(np.mean((sensor_data - rolling_median) ** 2))

def plot_sensor_data_with_global_rmse(column_name, sensor_dataframes, units_dict, window_size=10):
    """
    Plots data for the specified column from all sensor DataFrames, overlays a rolling median,
    and calculates a single RMSE for all sensors combined.
    """
    # Step 1: Aggregate data based on `time_stamp`
    df = aggregate_sensor_data(sensor_dataframes, column_name)

    # Step 2: Exclude `time_stamp` from rolling operation
    rolling_median = df.drop(columns=["time_stamp"]).rolling(window=window_size, min_periods=1).median()
    overall_median = rolling_median.median(axis=1)

    # Step 3: Calculate global RMSE
    all_sensor_data = df.drop(columns=["time_stamp"]).values.flatten()
    all_median_data = np.tile(overall_median.values, df.shape[1] - 1)  # Repeat median for each sensor
    global_rmse = calculate_rmse(all_sensor_data, all_median_data)

    # Step 4: Plot each sensor's data
    plt.figure(figsize=(10, 6))
    for sensor_name in df.columns[1:]:  # Skip 'time_stamp'
        plt.plot(df['time_stamp'], df[sensor_name], label=sensor_name)

    # Plot the rolling median
    plt.plot(df['time_stamp'], overall_median, color='black', linestyle='dashed', label='Rolling Median')

    # Add plot details
    plt.xlabel('time_stamp')
    plt.ylabel(f'{column_name} ({units_dict[column_name]})')
    plt.title(f'{column_name} Data for Sensors')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True)
    plt.tight_layout()  # Adjust layout to prevent clipping

    # Step 5: Annotate RMSE on the plot
    plt.text(0.05, 0.95, f'Global RMSE: {global_rmse:.2f}', fontsize=12,
             transform=plt.gca().transAxes, verticalalignment='top',
             bbox=dict(boxstyle='round', facecolor='white', alpha=0.7))

    # Show the plot
    plt.show()

    # Print global RMSE for reference
    print(f"Global RMSE: {global_rmse:.2f}")

# For individual Plot of columns
# Plot for 1-120 sensors
# plot_sensor_data_with_global_rmse('pm2.5_alt', cleaned_sensor_dataframes_1_120, units_dict, window_size=10)

# # Plot for 121-206 (06.04-06.06) sensors
# plot_sensor_data_with_global_rmse('pm2.5_alt', cleaned_sensor_dataframes_121_206_04_06, units_dict, window_size=10)

# # Plot for 121-206 (06.06-06.07) sensors
# plot_sensor_data_with_global_rmse('pm2.5_alt', cleaned_sensor_dataframes_121_206_06_07, units_dict, window_size=10)
```

```python
from sklearn.linear_model import LinearRegression

# Function to fit each sensor's data to the linear function
def fit_sensor_to_median(x, y):
    lr = LinearRegression()
    lr.fit(x, y)
    return lr.coef_[0][0], lr.intercept_[0]  # Return the coefficients and intercept of the linear regression model

def fit_sensor_data(column_name, sensor_dataframes, units_dict, window_size=10):
    """
    Fits sensor data to a rolling median and computes coefficients and intercepts.
    """
    # Aggregate sensor data based on `time_stamp`
    df = aggregate_sensor_data(sensor_dataframes, column_name)

    # Exclude `time_stamp` for rolling operations
    rolling_median = df.drop(columns=['time_stamp']).rolling(window=window_size, min_periods=1).median()

    fitted_data_dict = {}
    coefficients_data = []

    # Iterate over each sensor's column
    for sensor_name in df.columns[1:]:  # Skip `time_stamp`
        # Use the rolling median as the target
        X = df[sensor_name].to_frame(name=column_name)
        Y = rolling_median.median(axis=1).to_frame('median')

        # Fit data to the rolling median
        coefficients, intercept = fit_sensor_to_median(X, Y)
        fitted_data = X * coefficients + intercept
        fitted_data = fitted_data.values.flatten()

        # Store fitted data and coefficients
        fitted_data_dict[sensor_name] = fitted_data
        coefficients_data.append({
            'Sensor': sensor_name,
            'Coefficient': coefficients,
            'Intercept': intercept
        })
    
    # Create a DataFrame to store coefficients and intercepts
    coefficients_df = pd.DataFrame(coefficients_data)

    return coefficients_df
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def compare_harmonized_vs_non_harmonized(column_name, non_harmonized_df, harmonized_df, window_size=10):
    """
    Compare the median of non-harmonized vs harmonized data and calculate RMSE for both.
    """
    # Step 1: Calculate rolling median for both datasets
    rolling_median_non_harmonized = non_harmonized_df.drop(columns=["time_stamp"]).rolling(window=window_size, min_periods=1).median()
    rolling_median_harmonized = harmonized_df.drop(columns=["time_stamp"]).rolling(window=window_size, min_periods=1).median()

    # Step 2: Calculate overall median for each
    overall_median_non_harmonized = rolling_median_non_harmonized.median(axis=1)
    overall_median_harmonized = rolling_median_harmonized.median(axis=1)

    # Step 3: Calculate RMSE for each dataset
    all_sensor_data_non_harmonized = non_harmonized_df.drop(columns=["time_stamp"]).values.flatten()
    all_sensor_data_harmonized = harmonized_df.drop(columns=["time_stamp"]).values.flatten()

    all_median_data_non_harmonized = np.tile(overall_median_non_harmonized.values, non_harmonized_df.shape[1] - 1)
    all_median_data_harmonized = np.tile(overall_median_harmonized.values, harmonized_df.shape[1] - 1)

    rmse_non_harmonized = np.sqrt(np.mean((all_sensor_data_non_harmonized - all_median_data_non_harmonized) ** 2))
    rmse_harmonized = np.sqrt(np.mean((all_sensor_data_harmonized - all_median_data_harmonized) ** 2))

    # Step 4: Plot comparison
    plt.figure(figsize=(12, 6))

    # Non-harmonized data plot
    plt.plot(non_harmonized_df['time_stamp'], overall_median_non_harmonized, label='Non-Harmonized Median', color='blue', linestyle='--')

    # Harmonized data plot
    plt.plot(harmonized_df['time_stamp'], overall_median_harmonized, label='Harmonized Median', color='green', linestyle='--')

    # Add RMSE annotations
    plt.text(0.05, 0.95, f'Non-Harmonized RMSE: {rmse_non_harmonized:.2f}', fontsize=12, transform=plt.gca().transAxes, verticalalignment='top', color='blue')
    plt.text(0.05, 0.90, f'Harmonized RMSE: {rmse_harmonized:.2f}', fontsize=12, transform=plt.gca().transAxes, verticalalignment='top', color='green')

    # Add details
    plt.xlabel('time_stamp')
    plt.ylabel(f'{column_name}')
    plt.title(f'Comparison of Non-Harmonized vs Harmonized Data ({column_name})')
    plt.legend()
    plt.xticks(rotation=45, ha='right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Print RMSE values
    print(f"Non-Harmonized RMSE: {rmse_non_harmonized:.2f}")
    print(f"Harmonized RMSE: {rmse_harmonized:.2f}")

def harmonize_sensor_data(sensor_dataframes, coefficients_df, column_name):
    """
    Harmonizes sensor data using coefficients and intercepts from a DataFrame.
    
    Parameters:
    - sensor_dataframes: Dictionary of sensor DataFrames (key: sensor name).
    - coefficients_df: DataFrame with `Sensor`, `Coefficient`, and `Intercept`.
    - column_name: The column to harmonize (e.g., 'pm2.5_alt').
    
    Returns:
    - harmonized_dataframes: Dictionary of harmonized sensor DataFrames.
    """
    harmonized_dataframes = {}

    for sensor_name, sensor_df in sensor_dataframes.items():
        # Get the coefficient and intercept for the current sensor
        row = coefficients_df[coefficients_df['Sensor'] == sensor_name]
        if not row.empty:
            coefficient = row['Coefficient'].values[0]
            intercept = row['Intercept'].values[0]

            # Apply the harmonization formula
            harmonized_df = sensor_df.copy()
            harmonized_df[column_name] = harmonized_df[column_name] * coefficient + intercept
            harmonized_dataframes[sensor_name] = harmonized_df
        else:
            print(f"Warning: No coefficients found for sensor {sensor_name}. Skipping harmonization.")
    
    return harmonized_dataframes

```
